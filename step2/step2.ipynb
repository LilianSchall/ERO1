```python
import osmnx as ox
import networkx as nx
import matplotlib
import matplotlib.pyplot as plt
```

```python
def get_graph(name):
    G = ox.graph_from_place(name, network_type='all')
    G = ox.project_graph(G)
    return G
```

```python
def get_adjacency_lists(G):
    nodes = [];
    for e in G.nodes():
        nodes.append(e);
    l = len(nodes)
    adj = []
    for i in range(l):
        adj.append([])
    for u, v, length in G.edges(data="length", keys=False):
        if (u < v): #avoid getting edges in both ways (FIXME: Directed roads)
            continue
        iu = nodes.index(u)
        iv = nodes.index(v)
        visited = {"snow": False} #pointer like in python
        adj[iu].append((iv, round(length, 2), visited))
        adj[iv].append((iu, round(length, 2), visited))
    return adj
```

```python
outremont = get_graph("Outremont, Montreal, Canada")
```

```python
g = get_adjacency_lists(outremont)
for i in range(len(g)):
    print(i, g[i])
```
